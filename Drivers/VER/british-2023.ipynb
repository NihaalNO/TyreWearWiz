## Importing Packages

```python
import fastf1
import fastf1.plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fastf1.plotting.setup_mpl()
```

## Loading Session Data

```python
year = 2023
race = "British Grand Prix"

session = fastf1.get_session(year, race, 'R')
session.load()
```

## Load and Prepare Lap Data

```python
laps = session.laps

laps = laps[[
    'Driver', 'LapNumber', 'LapTime',
    'Compound', 'TyreLife', 'Stint', 'Team'
]].dropna()

laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
laps.head()
```

## Tyre Degradation Analysis on different Tyre Compounds

```python
driver = 'VER'
driver_laps = laps[laps['Driver'] == driver]

plt.figure(figsize=(12,6))

for compound in ['SOFT', 'MEDIUM', 'HARD']:
    data = driver_laps[driver_laps['Compound'] == compound]
    plt.plot(
        data['TyreLife'],
        data['LapTimeSeconds'],
        marker='o',
        label=compound
    )

plt.xlabel("Tyre Life (Laps)")
plt.ylabel("Lap Time (s)")
plt.title(f"Tyre Degradation  {driver}")
plt.legend()
plt.grid(True)
plt.show()
```

## Tyre Degradation Per Compound

```python

from numpy.polynomial.polynomial import polyfit

for compound in driver_laps['Compound'].unique():
    data = driver_laps[driver_laps['Compound'] == compound]
    if len(data) > 5:
        b, m = polyfit(data['TyreLife'], data['LapTimeSeconds'], 1)
        print(f"{compound}: +{m:.3f} sec/lap degradation")
```
